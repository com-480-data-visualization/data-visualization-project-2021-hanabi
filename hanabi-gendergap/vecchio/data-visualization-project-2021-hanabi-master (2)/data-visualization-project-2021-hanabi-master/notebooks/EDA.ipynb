# Data Exploration

```python
import pandas as pd
```

## Preprocessing on our data

```python
df = pd.read_csv("../data/data.csv")
df.head(5)
```

We would like to clean up our data by dropping rows with NaN

```python
# drop row with NaN
i=2006
while i<=2020:
    df.dropna(subset = [str(i)], inplace=True)
    i+=1
```

```python
df.head(5)
```

Exporting 3 dataframe, data with no NAN, data with only Index, data with only Rank

```python
df1 = df[df['Subindicator Type'] == 'Index']
df2 = df[df['Subindicator Type'] == 'Rank']
```

```python
# show data contains only index
df1.head(5)
```

```python
# show data contains only rank
df2.head(5)
```

```python
# export data
df.to_csv (r'../data/data_clean.csv', index = False, header=True)
df1.to_csv (r'../data/data_index.csv', index = False, header=True)
df2.to_csv (r'../data/data_rank.csv', index = False, header=True)
```

## Diving deep into data

```python
uniq = df['Indicator'].unique()
print("There are "+ str(len(uniq)) + " indicators in our data: ")
print(uniq)
```

```python
uniq2 = df['Country Name'].unique()
print("There are "+ str(len(uniq2)) + " countries in our data: ")
print(uniq2)
```

```python
grouped_df = df1.groupby("Indicator")
grouped_df = grouped_df.agg({"Country Name": "nunique"})
grouped_df = grouped_df.reset_index()
grouped_df = grouped_df.rename(columns={'Country Name': 'Country Count'})
grouped_df = grouped_df.sort_values('Country Count')
grouped_df
```

From the above table, we know that each indicator contains 103 country data.

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/data.csv")
df.head(5)
```

```python
uniq = df['Indicator'].unique()
uniq[:5]
```

```python
df.head(5)
```

```python
# drop row with NaN
i=2006
while i<=2020:
    df.dropna(subset = [str(i)], inplace=True)
    i+=1
```

```python
df.head(5)
```

```python
df = df[df["Subindicator Type"] != "Rank"]
df.head(5)
```

```python
grouped_df = df.groupby("Indicator")
grouped_df = grouped_df.agg({"Country Name": "nunique"})
grouped_df = grouped_df.reset_index()
grouped_df = grouped_df.sort_values('Country Name')
grouped_df
```

```python
grouped_df.to_csv (r'indicator_count.csv', index = False, header=True)
```
